#### Ejemplo de if

```python
# Ejemplo 1: Clasificación de un número (Uso de if-else y operador de módulo)
# Este código determina si un número ingresado es par o impar

numero = int(input("Ingresa un número: ")) # 1. Solicita entrada al usuario y la convierte a entero.
if numero % 2 == 0:                            # 2. Evalúa si el número es par usando el operador de módulo.
    print("El número es par.")                 # 3. Se ejecuta si el número es par.
```

ejemplo de else

```python
# Ejemplo 2: Verificación de Rango (Uso de operadores lógicos)
# Verifica si un número está dentro de un rango específico (entre 10 y 20 inclusive)

numero = int(input("Ingresa un número: ")) # 1. Captura el número del usuario.
if numero >= 10 and numero <= 20:          # 2. Evalúa si cumple ambas condiciones al mismo tiempo.
    print("El número está en el rango.")   # 3. Se ejecuta si está entre 10 y 20.
else:                                      # 4. De lo contrario.
    print("El numero no esta en el rango.")# 5. Informa que no cumple la doble condición.
```

```python
# Ejemplo 3: Clasificación de un número (Uso de if-elif-else y operadores relacionales)
# Este código determina si un número ingresado es positivo, negativo o cero
numero = int(input("Ingresa un número: ")) # 1. Solicita entrada al usuario y la convierte a entero.
if numero > 0:                             # 2. Evalúa si el número es mayor a cero.
    print("El número es positivo.")        # 3. Se ejecuta si la línea 2 es verdadera.
elif numero < 0:                           # 4. Si la línea 2 fue falsa, evalúa si es menor a cero.
    print("El número es negativo.")        # 5. Se ejecuta si la línea 4 es verdadera.
else:                                      # 6. Si las condiciones anteriores fallaron, entra aquí.
    print("El número es cero.")            # 7. Se ejecuta como última opción (el número es 0).
```

```python
# Ejemplo 4: Cálculo de Descuento (Uso de operadores aritméticos y relacionales)
# Determina un porcentaje de descuento basado en la cantidad de productos comprados

cantidad = int(input("¿Cuántos artículos? ")) # 1. Recibe la cantidad de artículos comprados.
if cantidad >= 10:                             # 2. Compara si la cantidad es mayor o igual a 10.
    descuento = 0.2                            # 3. Asigna 20% de descuento si es verdad.
elif cantidad >= 5:                            # 4. Si es menor a 10 pero mayor o igual a 5.
    descuento = 0.1                            # 5. Asigna 10% de descuento.
else:                                          # 6. Si compró menos de 5 artículos.
    descuento = 0                              # 7. No hay descuento.
# 8. Muestra el porcentaje multiplicando el valor decimal por 100.
print(f"Tu descuento es del {descuento * 100}%.")
```
